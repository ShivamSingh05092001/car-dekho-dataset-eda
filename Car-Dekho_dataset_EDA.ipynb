```python

def brand_model_analysis(brand):
    temp_df = df1.groupby(['brand','model'])['selling_price'].agg(['min','max','mean','count']).reset_index().round(2)
    ans = temp_df[temp_df['brand'] == brand]
    print(f'''
    Brand {brand} Stats:
    Total Cars  : {ans['count'].sum()}
    Total Models: {ans['model'].count()}
    ''')
    return ans
```

```python
brand_model_analysis('BMW')
```

```python
# 6.2 Dynamic Analysis by using Function 
def brand_model_analysis(brand):
    temp_df = df1.groupby(['brand', 'model'])['selling_price'].agg(['min', 'max', 'mean', 'count']).reset_index().round(2)
    
    # Filtering for the specific brand provided in the function argument
    ans = temp_df[temp_df['brand'] == brand]
    
    # Finding the model with the highest and lowest prices
    highest_price_model = ans[ans['max'] == ans['max'].max()]['model'].values[0]
    lowest_price_model = ans[ans['min'] == ans['min'].min()]['model'].values[0]
    
    # Getting the actual price values
    highest_price = ans[ans['max'] == ans['max'].max()]['max'].values[0]
    lowest_price = ans[ans['min'] == ans['min'].min()]['min'].values[0]
    
    # Printing the analysis results
    print(f'''
    Brand {brand} Stats:
    Total Cars: {ans['count'].sum()}
    Total Models: {ans['model'].count()}
    Highest Model: {brand}:{highest_price_model} | Price : {highest_price}
    Lowest Model: {brand}:{lowest_price_model} | Price : {lowest_price}
    ''')
    
    return ans
```

```python
brand_model_analysis('Toyota')
```

##  Insights

1. **Brand Dominance**
   - Maruti and Hyundai have the highest number of listings across all charts.
   - This indicates strong demand and high resale activity for these brands in the used car market.

2. **Model-wise Trends**
   - Models such as Swift, Alto, i20, City, and Verna appear most frequently.
   - These models are commonly used for daily commuting and family purposes.

3. **Transmission Preference**
   - Manual transmission vehicles are significantly more common than automatic vehicles.
   - Automatic transmission is mainly concentrated in premium brands like BMW, Audi, and Mercedes.

4. **Seller Type Distribution**
   - Dealer-listed vehicles dominate the market compared to individual sellers.
   - Trustmark dealers contribute a smaller but more reliable segment of listings.

5. **Mileage Distribution**
   - Most vehicles fall within the mileage range of **1525 km/l**.
   - This shows that fuel efficiency is a key consideration for buyers.

6. **Engine Capacity Distribution**
   - A majority of cars have engine capacities between **1000cc and 1500cc**.
   - Higher engine capacities are mostly associated with luxury brands and automatic transmission.

7. **Brand vs Transmission Relationship**
   - Mass-market brands mostly offer manual transmission vehicles.
   - Premium brands show a higher proportion of automatic vehicles despite lower overall counts.

8. **Market Structure**
   - The used car market is clearly skewed towards affordable, practical, and fuel-efficient vehicles.

9. **Resale-Oriented Models**
   - High-frequency models show strong resale potential due to easy availability of spare parts
     and lower maintenance costs.

10. **Urban Usage Pattern**
    - Cars with moderate engine size and good mileage are more common,
      indicating higher usage in city and urban driving conditions.


##  Outcomes and Suggestions

This section summarizes the key results of the analysis and provides
practical recommendations based on observed market patterns.

###  Outcomes
- The used car market is strongly influenced by **budget-friendly and reliable brands**.
- **Manual transmission vehicles** remain the preferred choice for a majority of buyers.
- **Dealer-based sales** offer wider market coverage and enhance customer trust.
- **Premium cars** form a niche segment, typically featuring higher engine capacity
  and greater adoption of automatic transmission.

###  Suggestions
- Dealers should prioritize **high-demand brands** such as Maruti and Hyundai
  to achieve faster inventory turnover.
- **Fuel-efficient vehicles** should be clearly highlighted in listings
  to attract cost-conscious buyers.
- The inventory of **automatic transmission cars** can be gradually expanded,
  particularly in metro and urban markets.
- Online platforms should actively promote **Trustmark and verified dealers**
  to strengthen buyer confidence.
- Buyers seeking daily-use vehicles should focus on cars with
  **moderate engine capacity (10001500cc)** and **good mileage** to reduce ownership costs.
